## 1 Important concepts

### 1.1 Path

The path is a method to specify a location on the computer. It is like an address, and if you follow the paht, it will take you to your file or folder.

Just like specifying location, you can specify your path **absolutely** or **relatively**. For example, specifying 'SPS is located on level 3 of block S16' is an absolute path, while if I am already on level 5 of S16, and I say, 'go two floors down' is a relative path. 

Here is an example of an absolute path to a file on the Desktop on a Windows machine

C:\\Users\Chammika\Desktop\data-01.txt

### 1.2 More about relative paths

When dealing with relative paths, it is helpful to know . and .. notation. For example, we use . to say 'this folder' and .. to say 'one folder above'.

- .\date-files\date-01.txt means the file data-01.txt in the folder data-files in the **current** folder
- ..\data-files\data-01.txt means the file data-01.txt in the folder data-files located in the folder **above**

Mac OS X and Linux allow you to use ~ to refer to your home directory. For example, you can access the Desktop in these systems 'relatively' with ~/Desktop

### 1.3 Path separator

Major OSs (Windows, Mac OS X, Linux) offer similar grpahical environments. One of the most striking differences between Windows and Mac OS X (or Linux) is the path separator.

Windows uses \ as the path separator while Mac OS X (or Linux) uses /. 

The path to a file on the Desktop on each of these systems will look like:

|  Operating Systems  | Path                                     |
|:-------------------:|------------------------------------------|
|       Windows       | C: \\ Users\chammika\Desktop\data-01.txt |
| Mac OS X (or Linux) | /Users/chammika/Desktop/data-01.txt      |


### Text files vs Binary files

All files on your computer can be thought of as **text files** or **binary files**. Text files are simple and can be opened, and their contents are examined by any software (e.g. Notepad, TextEdit, Jupiter,…). Examples of text file formats are .txt, .md or .csv.

Binary files require some processing to make sense of what they contain. For e.g. if you look at the raw data in a .png file, you will see gibberish. Furthermore, some binary files will only run on specific OSs. For example, the Excel.app on a Mac will not run on Windows, nor will the Excel.exe file run on Mac OS X (or Linux). Some reasons for having binary files are speed and size; text files, though simple, can get bulky.


### Extentions

Files are usually named to end with an extension separated from the name by a . like name.extension. This extension lets the OS know what software or app to extract the details in a file. For example, a .xlsx means use Excel or .pptx mean use PowerPoint. Changing the extension of a file inappropriately will make your OS throw a fit. 


### 1.4 Opening and closing files

An advanced but better way of opening a file for reading and writing is by using the with statement (called a **context manager**).

```python
#Typically, to read a file:

with open('spectrum-01.txt', 'r') as file:
    file_content = file.read()

print(file_content)

#The open() function opens a ‘stream’ into your file. The 'r' specifies that I only want to read from the file. Using with frees you from worrying about closing the file stream after you are done.
```

```python
#To write in a file, 
text = 'Far out in the uncharted backwaters of the unfashionable end of the western spiral arm of the Galaxy lies a small unregarded yellow sun.\nOrbiting this at a distance of roughly ninety-two million miles is an utterly insignificant little blue green planet whose ape-descended life forms are so amazingly primitive that they still think digital watches are a pretty neat idea.'

#Method 1: Writing everything in one go

with open('my-text-once.txt', 'w') as file:
    file.write(text)
    
#the file my-text-once.txt is now in my directory and contains text
#the 'w' indicates opening the file stream for writing

#Method 2: Write a line at a time
#Useful for writing data that is being generated on the fly
#Don't have such data now, so will split the lines of the previous text 

with open('my-text-lines.txt', 'w') as file:
    for line in text.splitlines():
        file.writelines(line)
        
#Writing to a file is a relatively slow operation. Will slow things down if you do it in a loop
```

## 2 Playing with the OS

### 2.1 Some packages

To programmatically create, copy, and delete files and folders navigate the OS, the following three packages can be used.

| Package | Primarily used for                                                                |
|---------|-----------------------------------------------------------------------------------|
| os      | To ‘talk’ to the OS to create, modify, delete folders and write OS-agnostic code. |
| glob    | To search for files                                                               |
| shutil  | To copy files.                                                                    |

Both os and shutil packages are used because shutil offers some function (e.g. shutil.copy()) that os does not. There are also subtile differences in what these functions do.

```python
#Import the packages
import os
import glob
import shutil
```

### 2.3 Creating folders

```python
#One can create a folder programmatically using os.mkdir()

os.mkdir('people')

for person in ['Jonh', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    print(f'Creating {path}')
    os.mkdir(path)
```

### 2.4 Checking for existence

```python
#Python will complain if you run the above code twice, saying that the file (Python referes folders as files) already exists.
#Need to check if they exist before creating resources
#Two ways to do this: try-except with the FileExistsError or os.path.exists()

#Using try-except
for person in ['Jonh', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    try:
        os.mkdir(path)
        print(f'Creating {path}')
    except FileExistsError:
        print(f'{path} already exists; skipping creation.')
```

```python
#Using os.path.exists()
for person in ['Jonh', 'Paul', 'Ringo']:
    path = os.path.join('people', person)
    if os.path.exists(path):
        print(f'{path} already exists; skipping creation.')
    else:
        os.mkdir(path)
        print(f'Creating {path}')
```

### 2.5 Copying files

```python
#to copy files programatically, using an example of copying the 73 logo (sp2273_logo,png) into each of the folders just created
for person in ['Jonh', 'Paul', 'Ringo']:
    path_to_destination = os.path.join('people', person)
    shutil.copy('sp2273_logo.png', path_to_destination)
    print(f'Copied file to {path_to_destination}')
```

```python
#If I want to have all the images in a folder called imgs in each person's directory, I can first create the folders imgs and then moving the logo file into that folder

for person in ['Jonh', 'Paul', 'Ringo']:
    # Create folder 'imgs'
    path_to_imgs = os.path.join('people', person, 'imgs')
    if not os.path.exists(path_to_imgs):
        os.mkdir(path_to_imgs)

    # Move logo file
    current_path_of_logo = os.path.join('people', person, 'sp2273_logo.png')
    new_path_of_logo = os.path.join('people', person, 'imgs', 'sp2273_logo.png')

    shutil.move(current_path_of_logo, new_path_of_logo)
    print(f'Moved logo to {new_path_of_logo}')
```

### 2.6 Listing files

```python
#to know what files are in a folder, glob does it easily
#if i want all the files in the current directory:
glob.glob('*')
```

The * is called a **wildcard** and is read as **'anything'**. so the above code is asking glob to give me anything in the folder

```python
#if i want to refine my search to give the files that match the pattern 'peo' followed by 'anything':
glob.glob('peo*')
```

```python
#if i want to know what is inside the folder 'peo' followed by 'anything':
glob.glob('peo*/*')
```

```python
#to see the whole, detailed structure of the folder people,
#need to tell glob to search recursively (i.e. dig through all sub-file directories) by putting recursive = True, and using two wildcards like this:
glob.glob('people/**', recursive=True)
```

```python
#if i only want to find .png files, i just need to modify it:
glob.glob('people/**/*.png', recursive=True)
```

## Exercise 8 :  Tidying a collaboration

```python
def rename_my_file(old_file_name):
    month = {
            'jan': '01', 'feb': '02', 'mar': '03',
            'apr': '04', 'may': '05', 'jun': '06',
            'jul': '07', 'aug': '08', 'sep': '09',
            'oct': '10', 'nov': '11', 'dec': '12'
            }
    
    file_directory = old_file_name.split(os.path.sep)[0]
    file_name = old_file_name.split(os.path.sep)[-1]
    file_month = file_name.split('-')[0]
    rest_of_file = file_name.split('-')[1]
    if file_month in month:
        if os.path.sep in old_file_name:
            new_file = file_directory + os.path.sep + month[file_month] + '-' + rest_of_file
            return new_file
        else:
            new_file = month[file_month] + '-' + rest_of_file
            return new_file
    else:
        return old_file_name
```

```python
for file in ['oct-08_singapore.txt', 'jul-10_cairo.txt', 'may-15_dhaka.txt',
 'may-13_cairo.txt', 'oct-21_cairo.txt', 'jan-10_singapore.txt',
 'jun-20_tokyo.txt', 'aug-06_jakarta.txt', 'dec-21_karachi.txt',
 'jan-01_tokyo.txt']:
    new_file_name = rename_my_file(file)
    print(f'{file} ----> {new_file_name}')
```

```python
#to find only .txt files
for file in glob.glob('os-collaboration-exercise-data/**/*.txt', recursive=True):
    os.rename(file, rename_my_file(file))
```

```python
#change directory to os-collaboration-exercise-data
os.chdir('os-collaboration-exercise-data')
os.getcwd()
```

```python
#create folder for every city
cities = ['Cairo', 'Dhaka', 'Jakarta', 'Karachi', 'Manila',
 'Melbourne', 'Osaka', 'Shanghai', 'Singapore', 'Tokyo']

for city in cities:
    os.mkdir(city)
```

```python
#list all the files from Cairo
glob.glob('Cairo/**', recursive=True)
```

```python
#copying all the files from Cairo to Cairo folder
for file in glob.glob('**'):
    if 'cairo' in file:
        shutil.copy(file, 'Cairo')
    
```

```python
#List all files from Cairo
glob.glob('Cairo/**', recursive=True)
```

```python
#Move all the files to the corresponding folders of the cities
for file in glob.glob('**'):
    for city in cities:
        if city.lower() in file:
            new_path = os.path.join(city, file)
            if not os.path.exists(new_path):
                shutil.move(file, new_path)
```
