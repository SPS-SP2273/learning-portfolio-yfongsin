### 1.1 Letâ€™s look at some code.

```python
#To plot a curve

# Some data for plotting
x = [0, 1, 2, 3, 4, 5, 6]
y_1 = [0, 2, 4, 6, 8, 10, 12]
y_2 = [0, 3, 6, 9, 12, 15, 18]

# Let's start plotting
plt.plot(x, y_1, color='red', linestyle='dashed', label='Y values') #define your x and y a-xis, color of plot, linestyle, and label for line
plt.xlabel('x-values') #x-axis label
plt.ylabel('y-values') #y-axis label
plt.title('X vs Y') #graph title
plt.grid(alpha=.25) #grids and the boldness of the grid
plt.legend(loc='upper left') #where to put the legend
plt.show() #show the plotted graph
```

### 1.6 Adding mathematical functions

```python
#One advantage of NumPy arrays is that it allows us to easily generate data-relataed maths functions
#Reusing our previous code to plot x**2 and sin(x)

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend()
plt.show()
```

Plot does not look good because sin(x) lies between +-1 but x^2 has no such bounds. One way to fix this is to add another y-axis that shares the same x-axis

### 1.7 We need another axis!

```python
#matplotlib offers a variety of ways to have multiple axes
#Simplest way to have another y-axis that shares the same x-axis is to use the command twinx():

x = np.array([0, 1, 2, 3, 4, 5, 6])
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

**Note** that we have two legend() calls, one for each axis.

However, plot still does not look good because we have limited points. Can use np.linspace to fix this with:

```python
x = np.array([0, 1, 2, 3, 4, 5, 6])
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.show()
```

### 1.8 Saving to disc

To use a plot in a report or presentation, you must first save it to disk. matplotlib makes it easy to export plots into many formats (PDF, JPEG, PNG, BMP...) and at different resolutions.

To do this, we use the function savefig() and specify the format with the extension (e.g. filename.pdf) of the file name and resolution by specifying the dots-per-inch (dpi)

```python
x = np.linspace(0, 6, 100)
x2 = x**2
sin_x = np.sin(x)

plt.plot(x, x2, color='red', linestyle='dashed', label='x^2')
plt.legend(loc='lower left')

# This creates a new y-axis for the plot that comes after
plt.twinx()
plt.plot(x, sin_x, color='blue', label='sin(x)')
plt.legend(loc='lower right')

plt.savefig('simple-plot.png', dpi=150)

#This will save the file in the same directory as the one my notebook is in.
#Can specify the path in more detail for e.g.
#plt.savefig('~/Desktop/simple-plot.png', dpi=150)
```

## 2 A real example: Global Warming

### 2.1 Plotting data from files

```python
#Need to know how to plot data stored in a file (e.g. spreadsheet, text file , database)

#Using average temperature data for Daily Land in the Berkeley Earth website
#Loaading the txt file into data
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0] #accessing date from data
anomaly = data[:, -1] #accessing anomaly from data

#plot
plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])
```

```python
#To add a horizontal line at the zero value to highlight the trend shown by the data
#hlines() function needs a y-value and starting and ending values for x
data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)
date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.ylim([-8, 8])

plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
```

```python
#matplotlib offers some standard style templates

#to use the fivethirtyeight style template

plt.style.use('fivethirtyeight')

data = np.loadtxt('Complete_TAVG_daily.txt', skiprows=24)

date = data[:, 0]
anomaly = data[:, -1]

plt.plot(date, anomaly, alpha=.5)
plt.hlines(0, date[0], date[-1], linestyle='--', colors='grey')
plt.ylim([-8, 8])

plt.xlabel('Date')
plt.ylabel('Temperature Anomaly')
plt.title('Temperature anomaly\n(Relative to  average from Jan 1951 - Dec 1980.)')
plt.show()
```
